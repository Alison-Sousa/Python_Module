**1) A partir da base de dados notas.xlsx, criar um subconjunto de dados (novo dataSet) com os Alunos, Disciplinas e Médias apenas para a disciplina de Biologia, grave o resultado em um arquivo CSV chamado Biologia.csv**

```python
import pandas as pd 

tabela_prova = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Notas.xlsx')
colunas_selecionadas = ['Aluno', 'Disciplina', 'Média']
dados_selecionados = tabela_prova.filter(items=colunas_selecionadas)

dados_selecionados.Disciplina
tabela_nova = dados_selecionados[(dados_selecionados.Disciplina == 'Biologia')]
tabela_nova.to_csv("/content/drive/MyDrive/Colab Notebooks/Biologia.csv")
tabela_nova.head()
```
